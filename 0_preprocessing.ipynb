### 2. Review

```python
reviews = pd.read_csv('ReviewRaw.csv')
```

```python
reviews.head(3)
```

```python
# Số lượng sản phẩm
print('Số lượng sản phẩm:', reviews.shape[0])
```

```python
# Kiểu dữ liệu
reviews.info()
```

```python
# Bỏ dữ liệu trùng
print('Trước khi droping, records =', reviews.shape[0])
reviews.drop_duplicates(inplace=True)
print('Sau khi droping, records =', reviews.shape[0])
```

```python
# Kiểm tra dữ liệu null
reviews.isnull().any()
```

```python
m = []
for c in ['full_name', 'created_time']:
    m.append({
        'feature':c,
        'MissingVal':reviews[reviews[c].isnull()].shape[0],
        'Percentage':reviews[reviews[c].isnull()].shape[0]/reviews.shape[0]*100
    })
df = pd.DataFrame(m)
df
```

#### Nhận xét:
- created_time feature có dữ liệu null khá lớn.
- full_name feature có dữ liệu null tương đối 9.5%.
- name và full_name(nếu có) khá giống nhau.

```python
# Bỏ feature full_name và created_time
reviews.drop(['full_name', 'created_time'], axis=1, inplace=True)
```

```python
# lưu kết quả
reviews.to_csv('Reviews.csv', index=False)
```

### 3. Product - Review

```python
# sản phẩm không có trong review
reviews[~reviews.product_id.isin(products.item_id)]
```

```python
# loại bỏ review có mã sản phẩm không tồn tại
reviews = reviews[reviews.product_id.isin(products.item_id)]
```

```python
reviews.reset_index(drop=True, inplace=True)
```

```python
# lưu kết quả
reviews.to_csv('Reviews.csv', index_label='id')
products.to_csv('Products.csv', index=False)
```
